```
#Import dependencies 


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection
from sklearn import datasets
```

```
#Load the data 
iris=datasets.load_iris()
x=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```
y=pd.DataFrame(iris.target,columns=["Species"])
```

### **Split the the into 80% training and 20% testing datasets**

```
X_train,X_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2, random_state=0)
```

### **Create Decision tree classifier and train on training data**

```
dtc= DecisionTreeClassifier().fit(X_train,y_train)
```

### **Prediction on training data**

```
pred = dtc.predict(X_train)
pred
```

```
y_train
```

### **Predicion in testing data**

```
test_pred = dtc.predict(X_test)
test_pred
```

```
y_test
```

### **Checking precision, recall and f1-score**

```
from sklearn.metrics import classification_report
print(classification_report(test_pred,y_test))
```

### **Evaluating the decision trees using entropy**

```
classifier= DecisionTreeClassifier(criterion='entropy',random_state=0).fit(X_train,y_train)
```

```
classifier.score(X_test,y_test)
```

```
y_pred = classifier.predict(X_test)
y_pred
```

```
y_test
```
